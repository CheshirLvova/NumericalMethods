```python
import matplotlib.pyplot as plt           
import numpy as np
import math
from math import tan
```

## Метод Ейлера

```python
def F(X,Y):
    #return X**2 - Y*2#
    return Y + (math.e**X)/X
Xlist =[]
Ylist =[]
def Euler(x,x1,y,n):
    h=(x1-x)/n
    for i in range(1,n+1):
        F1=F(x,y)
        y+=F1*h
        x+=h
        Xlist.append(float(x))
        Ylist.append(y) 
        print(i," ",x," ",i," ",y," ",(tan(x)-x))
x=1.
x1=2.
n=10
y=1
print("Euler method")
Euler(x,x1,y,n)
```

## Метод Рунге-Кутта 3-Б порядку

```python
def F(X,Y):
    #return X**2 - Y*2#
    return Y + (math.e**X)/X
X1list =[]
Y1list =[]

def RungeKutta(x,x1,y,n):
    h=(x1-x)/n
    for i in range(1,n+1):
        F1=F(x,y)
        k1=h*F1
        F2=F(x + 1./3.*h,y + 1./3.*k1)
        k2=h*F2
        F3=F(x + 2./3.*h,y + 2./3.*k2)
        k3=h*F3
        y+=1./4.*k1 + 3./4.*k3
        x+=h 
        X1list.append(float(x))
        Y1list.append(y) 
        print(i," ",x," ",i," ",y," ",(tan(x)-x))
#x=0.
#x1=1.
#n=10
#y=1
x=1.
x1=2.
n=10
y=1
print("Runge-Kutta method")
RungeKutta(x,x1,y,n)
```
