Побудувати інтерполяційний многочлен Лагранжа та за його допомогою знайти наближене значення функції у вказаній точці. Вивести його графік. 

```python
import numpy as np
import matplotlib.pyplot as plt
#x=np.array([1.5,1.54,1.6,1.63,1.72, 1.52], dtype=float)
#y=np.array([3.87,3.92,3.95,4.,4.21, 3.91], dtype=float)
x=np.array([-2., -1., 0], dtype=float)
y=np.array([4., 1., 0], dtype=float)
def lagranz(x,y,t):
    z=0
    for j in range(len(y)):
        p1=1; p2=1
        for i in range(len(x)):
            if i==j:
                p1=p1*1; p2=p2*1   
            else: 
                p1=p1*(t-x[i])
                p2=p2*(x[j]-x[i])
        z=z+y[j]*p1/p2
    return z
#print("L(1.52) = 3.914321")
xnew=np.linspace(np.min(x),np.max(x),1000)
ynew=[lagranz(x,y,i) for i in xnew]
plt.plot(x,y,'o',xnew,ynew)
plt.grid(True)
plt.show()
```
